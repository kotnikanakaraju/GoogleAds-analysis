```python
data['Payments'] = pd.to_numeric(data['Payment ($)'], errors='coerce').fillna(0)
data['cost'] = pd.to_numeric(data['Cost ($)'], errors='coerce').fillna(0)
```

```python
data['RoAS']=pd.to_numeric(data['Payments']/data['cost'],errors='coerce').fillna(0)
```

```python
import matplotlib.pyplot as plt
# Assuming you have already read the dataset into a DataFrame called 'data'
# Replace 'column_name' with the actual column name where you calculated RoAS
data['RoAS'] = pd.to_numeric(data['Payments'] / data['cost'], errors='coerce').fillna(0)


# Assuming you have a 'date' column in your DataFrame that contains the dates of ad spends
plt.plot(data['Week'], data['RoAS'])
plt.xlabel('Week')
plt.ylabel('RoAS')
plt.title('Return on Ad Spends over Time')
plt.xticks(rotation=45)  # To rotate the date labels for better visibility
plt.grid(True)
plt.show()
```

```python
# For example, to get the average RoAS for each cohort date
average_roas_by_cohort = data.groupby('Week')['RoAS'].mean()

# Or to get the total clicks and impressions for each cohort date
cohort_metrics = data.groupby('Week').agg({'Clicks': 'sum', 'Impressions': 'sum'})
```

```python
import seaborn as sns
plt.hist(average_roas_by_cohort, bins=10)
plt.xlabel('Average RoAS')
plt.ylabel('Frequency')
plt.title('Histogram of Average RoAS by Week')
plt.grid(True)
plt.show()
```

```python
# Assuming 'Week' is a numeric column, if not, you might need to encode it to numeric values
heatmap_data = data.pivot_table(index='Week', columns='Ad group', values='RoAS', aggfunc='mean')

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.xlabel('Ad group')  # Replace 'Some_Column' with the column you want on the x-axis
plt.ylabel('Week')
plt.title('Heatmap of Average RoAS by Week and Some_Column')
plt.show()
```

```python
cohort_metrics.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Week')
plt.ylabel('Count')
plt.title('Total Clicks and Impressions by Week')
plt.grid(True)
plt.legend(title='Metrics', loc='upper left')
plt.show()
```

```python
cohort_metrics['Clicks'].plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage Distribution of Clicks by Week')
plt.show()
```

```python
plt.hist(cohort_metrics['Impressions'], bins=10)
plt.xlabel('Impressions')
plt.ylabel('Frequency')
plt.title('Histogram of Total Impressions by Week')
plt.grid(True)
plt.show()
```
